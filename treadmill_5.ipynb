```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load datasets
test_measure_df = pd.read_csv("test_measure.csv")  # Your test_measure dataset
subject_info_df = pd.read_csv("subject-info.csv")  # Your subject_info dataset

# Merge the two datasets based on a common identifier (e.g., 'ID_test')
# Replace 'ID_test' with the actual column name that identifies subjects in both datasets
merged_df = pd.merge(test_measure_df, subject_info_df, on="ID_test", how="inner")

# Ensure required columns exist in the merged dataset
required_cols = ["VO2", "Weight", "Temperature", "Humidity"]
merged_df = merged_df.dropna(subset=required_cols)  # Remove missing values

# Compute VO2 max (ml/kg/min)
merged_df["VO2 max (ml/kg/min)"] = (merged_df["VO2"] / merged_df["Weight"]) * 1000

# Determine median VO2 max
median_vo2 = merged_df["VO2 max (ml/kg/min)"].median()

# Classify Fitness Level based on median VO2 max
merged_df["Fitness Level"] = merged_df["VO2 max (ml/kg/min)"].apply(lambda x: "Low" if x < median_vo2 else "High")

# Select features for pairplot
features = ["Temperature", "Humidity", "VO2 max (ml/kg/min)", "Fitness Level"]

# Pairplot with hue based on Fitness Level
sns.pairplot(merged_df[features], hue="Fitness Level", diag_kind="kde", palette={"Low": "red", "High": "green"})

# Show the plot
plt.show()
```

```python
# Load subject-info and test_measure into data frames 
df_subject = pd.read_csv('subject-info.csv')  
df_testmeasure = pd.read_csv('test_measure.csv')  

# Count the number of tests for each participant
test_counts = df_testmeasure['ID'].value_counts()

# Identify the participant with the maximum number of tests
max_tests_participant = test_counts.idxmax()  # Get ID of participant with max tests
max_tests_count = test_counts.max()  # Get the highest test count

# Filter the data for this participant
subject_info_filtered = df_subject[df_subject['ID'] == max_tests_participant]
test_measure_filtered = df_testmeasure[df_testmeasure['ID'] == max_tests_participant]

# Merge the two datasets on 'ID' and 'ID_test'
merged_df = subject_info_filtered.merge(test_measure_filtered, on=['ID', 'ID_test'])

# Generate descriptive statistics for this participant's test data
summary_stats = merged_df[['time', 'Speed', 'HR', 'VO2', 'RR']].describe()

# Display results
print(f"Participant with the maximum number of tests: {max_tests_participant} ({max_tests_count} tests)\n")

print("Participant Details:")
print(subject_info_filtered.to_string(index=False))

print("\nSummary Statistics for Test Measurements:")
print(summary_stats)
```

```python
merged_df = pd.merge(subject_info_df,test_measure_df, how='inner',on=['ID','ID_test'])
merged_df.sort_values(by=['ID','ID_test','time'], inplace=True)
merged_df.reset_index()


# Creating a column in merged_df for phase categorization
merged_df['phase'] = 'Unknown'

# Function to categorize phases for each test
def categorize_phases(ID_test_data):
    # Warm-Up: Speed <= 5 km/h, initial walking phase
    ID_test_data.loc[ID_test_data['Speed'] <= 5, 'phase'] = 'Warm-Up'

    # GET Phase: Speed > 5 km/h, indicating the graded exercise test
    ID_test_data.loc[(ID_test_data['Speed'] > 5) & (ID_test_data['phase'] == 'Unknown'), 'phase'] = 'GET'
    
    # Maximal Effort: Identified when Speed reaches its maximum for the test
    max_speed = ID_test_data['Speed'].max()     # Get the maximum speed for the current test
    max_VO2 = ID_test_data[ID_test_data['Speed']==max_speed]['VO2'].max()     # Get the maximum VO2 for the current test
    time_at_max_VO2= ID_test_data[ID_test_data['VO2']==max_VO2]['time'].min()   # Get the time at maximum VO2
    
    ID_test_data.loc[(ID_test_data['Speed'] == max_speed) & (ID_test_data['VO2'] == max_VO2)& (ID_test_data['phase'] == 'GET'), 'phase'] = 'VO2_Max_GET'
    
    # Cool-Down: Speed < 5 km/h and time > time at max VO2 (indicating recovery)
    ID_test_data.loc[(ID_test_data['Speed'] <= 5) &  (ID_test_data['time'] > time_at_max_VO2) &  ((ID_test_data['phase'] == 'VO2_Max_GET') | (ID_test_data['phase'] == 'Warm-Up')), 'phase'] = 'Cool-Down'
    
    return ID_test_data

# Group by 'ID_test'
grouped = merged_df.groupby('ID_test')

# Apply the 'categorize_phases' function to each group
merged_df = grouped.apply(categorize_phases , include_groups=False)

merged_df= merged_df.reset_index(drop=False) 


# Athlete who had maximun GET time :

def athlete_with_max_GET_time(df):
    athlete_GET_time = df[df['phase']== 'GET'][['ID','time','phase']].reset_index()
    max_time = athlete_GET_time['time'].max()
    # Find the athlete who had the maximum GET time
    athlete_max_time= athlete_GET_time.loc[athlete_GET_time['time']==max_time,'ID']    
    return athlete_max_time
    
athlete_max_time = athlete_with_max_GET_time(merged_df)

print("\n\nAthlete who had maximum GET time is \n\n",athlete_max_time.values[0])
```
